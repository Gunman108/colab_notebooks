##Pull all videos from channel id

```python
import urllib
import urllib.request
import json

def get_all_video_in_channel(channel_id):
    api_key = jump_apikey
    api_service_name = "youtube"
    api_version = "v3"

    base_video_url = 'https://www.youtube.com/watch?v='
    base_search_url = 'https://www.googleapis.com/youtube/v3/search?'

    first_url = base_search_url+'key={}&channelId={}&part=snippet,id&order=date&maxResults=25'.format(api_key, channel_id)

    youtube = googleapiclient.discovery.build(
        api_service_name, api_version, developerKey = api_key)

    video_data = []
    url = first_url
    while True:
        inp = urllib.request.urlopen(url)
        resp = json.load(inp)

        for i in resp['items']:
            if i['id']['kind'] == "youtube#video":
           
                url = base_video_url + i['id']['videoId']
                
                request = youtube.videos().list(
                  part="statistics",
                  id=i['id']['videoId']
                )
                response = request.execute()

                try:
                  views = response['items'][0]['statistics']['viewCount']
                except KeyError:
                  views = '0'
                try:
                  likes = response['items'][0]['statistics']['likeCount']
                except KeyError:
                  likes = '0'
  
                stats = [views, likes]
                thumnail = i['snippet']['thumbnails']['medium']['url']

                video_data.append({"url": url,"thumbnail": thumnail, "stats": stats})

        try:
            next_page_token = resp['nextPageToken']
            url = first_url + '&pageToken={}'.format(next_page_token)
        except:
            break
    return video_data
    
```

##Getting Subscribers from CSVs

```python
import json
import urllib.request
import string
import random
import csv
import pandas as pd
import glob
import os
import googleapiclient.discovery
```

```python
def get_subcount(channel_id):

    os.environ["OAUTHLIB_INSECURE_TRANSPORT"] = "1"

    api_service_name = "youtube"
    api_version = "v3"
    DEVELOPER_KEY = jump_apikey

    youtube = googleapiclient.discovery.build(
        api_service_name, api_version, developerKey = DEVELOPER_KEY)

    request = youtube.channels().list(
        part="snippet,contentDetails,statistics",
        id= channel_id
    )
    response = request.execute()

    try:
      subs = int(response['items'][0]['statistics']['subscriberCount'])
    except KeyError:
      subs = 0

    return subs
```
