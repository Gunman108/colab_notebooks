```python
import matplotlib.pyplot as plt 
import pandas as pd
from pandas import Categorical
import numpy as np
import numpy.random as random
import math
```

```python
species_data_directory = '/content/drive/MyDrive/CS/Datasets/species.csv'
park_data_directory = '/content/drive/MyDrive/CS/Datasets/parks.csv'
```

```python
species = pd.read_csv(species_data_directory)
park = pd.read_csv(park_data_directory)
```

Data preprocessing

```python
species.head()
```

```python
park.head(10)
```

```python
species.columns = [c.replace(' ', '_') for c in species.columns]
park.columns = [c.replace(' ', '_') for c in park.columns]
```

```python
species = species[species.Occurrence != "Not Confirmed"]
species = species[species["Occurrence"].str.contains("Not Present")==False]
species.head()
```

```python
species_group = species.groupby(['Park_Name']).count()
diversity = species_group["Species_ID"]
diversity.head()
```

```python
import statistics
diversity_list = diversity.tolist()
print(diversity_list)
average = statistics.mean(diversity_list)
median = statistics.median(diversity_list)
print("Median:", median)
print("Average:", average)
```

```python
catagory = []
for i in diversity_list:
  if i < average:
    catagory.append(0)
  else:
    catagory.append(1)
print(catagory)
#print counts of 0 and 1s
```

```python
park["Diversity_Catagory"] = catagory
park["Diversity"] = diversity_list
park.State = pd.Categorical(park.State)
park['State_Num'] = park.State.cat.codes
park.head()
```

X/y, normalization, and train/test splits

```python
#Xy
xdb = park[['Acres', 'Latitude', 'Longitude',"State_Num"]].copy()
ydb = park[['Diversity_Catagory']].copy()
X = np.array(xdb)  
y = np.array(ydb)
```

```python
#normalization
for column in xdb.columns:
    xdb[column] = (xdb[column] - xdb[column].min()) / (xdb[column].max() - xdb[column].min()) 
```

```python
#train/test
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.2,random_state=20, stratify=y)
```

```python
from keras.models import Sequential
from keras.layers import Dense, Dropout
from keras.callbacks import EarlyStopping
from keras.constraints import maxnorm
from sklearn.preprocessing import MinMaxScaler


num_cols = X_train.shape[1]

model = Sequential()

model.add(Dense(100, input_shape=(num_cols,), activation='relu'))
model.add(Dense(600, activation='relu'))
model.add(Dense(600, activation='relu'))
model.add(Dense(400, activation='relu'))
model.add(Dense(100, activation='relu'))
model.add(Dense(1, activation='sigmoid'))

print(model.summary())
```

```python
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

early_stopping_monitor = EarlyStopping(patience=40)

print(y_train.dtype)
print(X_train.shape)
print(y_train.shape)


model.fit(X_train, y_train, epochs=2000, validation_split=0.2, batch_size=128, callbacks=[early_stopping_monitor], validation_data=(X, y))
model.evaluate(X_test, y_test)
```
