Iterates through our table and assigns every position that we are not looking for with a 1 and the position we are looking for with a zero using a list

```python
lables = []
for index, row in nfl.iterrows(): 
    if (row["position"]=='OLB' or row["position"]=='K' or row["position"]=='P' or row["position"]=='RB' or row["position"]=='WR' or row["position"]=='FS' or row["position"]=='TE'or row["position"]=='CB' or row["position"]=='LS' or row["position"]=='SS' or row["position"]=='FB' or row["position"]=='QB' or row["position"]=='P') :
      lables.append(1)
    else:
      lables.append(0)
 
    

print(lables)
```

Creates a tables list based off of that iteration

```python
nfl['labels'] = lables
```

Graphs the same graph as we did before, but this time the color is based off of our 1 and 0 position values

```python
sns.countplot(x='position', hue='labels',palette='rainbow', data=nfl)
```

Creates a graph where every position is shown as either a 1 or a 0 on the x axis

```python
sns.set_style('whitegrid')
sns.lmplot('labels','weight',data=nfl, palette='coolwarm',hue='position',size=6,aspect=1,fit_reg=False)
```

Makes sure all datatypes are either floats or ints and creates a new table that drops anything else

```python
non_floats = []
for col in nfl:
    if (nfl[col].dtypes != "float64") and(nfl[col].dtypes != "int64"):
        non_floats.append(col)
newnfl = nfl.drop(columns=non_floats)
```

Prints our new table with dropped non floats and ints
