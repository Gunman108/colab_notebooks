```python
# sigmoid function
def nonlin(x,deriv=False):
  if(deriv==True):
    return x*(1-x) #getting derivative if specified
  return 1/(1+np.exp(-x)) #mapped to sigmoid
 
# input dataset
X = np.array(xdb) #input

print(X)
 
# output dataset
y = np.array(ydb)

# seed random numbers to make calculation deterministic (based only on input and inital conditions)
np.random.seed(4)

# initialize weights randomly with mean 0
syn0 = 2*np.random.random((4,1)) - 1

print("initial weights:",syn0)
epochs = 100000
for iter in range(epochs):
# forward propagation
  l0 = X
  l1 = nonlin(np.dot(l0,syn0)) #maps dot product of weights and input data to sigmoid
 

# how much did we miss?
  l1_error = y - l1 #how far off the prediction is

# multiply how much we missed by the slope of the sigmoid at the values in l1
  l1_delta = l1_error * nonlin(l1,True) #finds derivative at sigmoid

# update weights
  syn0 += np.dot(l0.T,l1_delta) #weights changed by error (slope of sigmoid error function)

  if(iter%200==0): #prints the loss after every 100 epochs
    print("epoch:",iter,"/",epochs,"|","Loss:",l1_delta)

print("Output After Training:")
print(l1)
print()
print("0 values:")
np.count_nonzero(l1 < .5)
```

```python
#Create newlabels
def layerlabels(y, split):
  newlabels = []
  for i in y:
    if i > split:
      newlabels.append(1)
    else:
      newlabels.append(0)
  return newlabels
```

```python
#create an accuracy checker that takes in labels and newlabels and finds false positives, negatives, 
def accuracycheck(labels,predictions):
  correct = 0
  false_positives = 0
  false_negatives = 0
  for j in zip(labels,predictions):
    if int(j[0]) == j[1]:
      correct += 1
    if int(j[0]) > j[1]:
      false_positives +=1
    if int(j[0]) < j[1]:
      false_negatives +=1

  accuracy = correct/(correct+false_positives+false_negatives)
  print("False Positives:",false_positives)
  print("False Negatives:",false_negatives)
  print("Correct:",correct)
  print()
  print("Accuracy:", accuracy)
  return accuracy
```

Corrected two layer network

```python
import numpy as np

def nonlin(x,deriv=False):
	if(deriv==True):
	    return x*(1-x)

	return 1/(1+np.exp(-x))
    
X = np.array(xdb)
                
y = np.array(ydb)

np.random.seed(6)

# randomly initialize our weights with mean 0
syn0 = 2*np.random.random((3,4)) - 1
syn1 = 2*np.random.random((4,1)) - 1

print("initial weights:",syn0)
epochs = 60000000

for j in range(epochs):

	# Feed forward through layers 0, 1, and 2
    l0 = X
    l1 = nonlin(np.dot(l0,syn0))
    l2 = nonlin(np.dot(l1,syn1))

    # how much did we miss the target value?
    l2_error = y - l2
    
    if(j%5000==0): #prints the loss after every 100 epochs
      print("epoch:",j,"/",epochs,"|","Loss:",str(np.mean(np.abs(l2_error))))
        
    l2_delta = l2_error*nonlin(l2,True)
    l1_error = l2_delta.dot(syn1.T)

    l1_delta = l1_error * nonlin(l1,True)

    syn1 += l1.T.dot(l2_delta)
    syn0 += l0.T.dot(l1_delta)

print("Output After Training:")
print(l2)
print()
print("0 values:")
np.count_nonzero(l2 < .5)
```

```python
from keras.models import Sequential
from keras.layers import Dense, Dropout
from keras.callbacks import EarlyStopping
from keras.constraints import maxnorm
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler



X = np.array(xdb)
                
y = np.array(ydb)

X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.2,random_state=20, stratify=y)

num_cols = X_train.shape[1]

model = Sequential()

model.add(Dense(60, input_shape=(num_cols,), activation='relu'))
model.add(Dense(60, activation='relu'))
model.add(Dense(120, activation='relu'))
model.add(Dense(60, activation='relu'))
model.add(Dense(1, activation='sigmoid'))

print(model.summary())
```

```python
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

early_stopping_monitor = EarlyStopping(patience=40)

print(X_train,y_train)

model.fit(X_train, y_train, epochs=2000, validation_split=0.3, batch_size=128, callbacks=[early_stopping_monitor], validation_data=(X_test, y_test))
model.evaluate(X_test, y_test)
```
