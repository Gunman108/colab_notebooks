```python
import numpy as np
import matplotlib.pyplot as plt
from numpy import random
import random

#Epoch, lr, reps
epoch = 1000000
lr = .01
num_points = 100

point_yrange = 200 
point_xrange = 200
m, b = 2,1 #what is the slope/y-int deviding the points
overlap_factor = 0 #how neatly seperated are the points

#Creating points
xpr = []
ypr = []

xpb = []
ypb = []

xpb = [random.randrange(start=0, stop=point_xrange) for i in range(num_points)]
xpr = [random.randrange(start=0, stop=point_xrange) for i in range(num_points)]

ypb = [random.randrange(start=int(m*x+b)-point_yrange, stop=int(m*x+b)+overlap_factor) for x in xpb]
ypr = [random.randrange(start=int(m*x+b)-overlap_factor, stop=int(m*x+b)+point_yrange) for x in xpr]

#creates a, b, and c for the function ax + by + c = 0

a= random.randint(-10, 10)
b= random.randint(-100, 100)

while b == 0: #Incase b becomes 0
  b = random.randint(-100, 100)

c= random.randint(0,100)



myy = -(a*200)/b - c/b
xp2 = np.array([0,200])
yp2 = np.array([b, myy])
plt.plot(xp2,yp2, label="inital line")

print('init a:',a)
print('init b:',b)
print('init c:',c)

bluecount = 0
redcount = 0

for i in range(epoch):
  rp = random.randint(0,99)
  binary = random.randint(0,2)
  if binary == 0: #red
    x = xpr[rp]
    y = ypr[rp]

  if binary == 1: #blue
    x = xpb[rp]
    y = ypb[rp]


  if((a*x)+(b*y)+ c > 0 and binary == 0): #positive region > 0
    redcount += 1
    a -= x*lr
    b -= y*lr
    c -= lr

  elif((a*x)+(b*y) + c < 0 and binary == 1): #negative region < 0
    bluecount += 1
    a += x*lr
    b += y*lr
    c += lr
      
print('new a:',a)
print('new b:',b)
print('new c:',c)
print('red points:',redcount)
print('blue points:', bluecount)
#Graph Points
plt.plot(xpr, ypr, 'o', color='red')
plt.plot(xpb, ypb, 'o', color='blue')


#graphs line
myy = -(a*200)/b - c/b
xp2 = np.array([0,200])
yp2 = np.array([b, myy])
plt.plot(xp2,yp2, color="purple", label="final line")


plt.legend(loc="upper left")
# plt.plot(ypb, ypb, 'o', color='blue')
plt.show()
```
