#Graph Analysis

Human Development Index

```python
sns.set_style('whitegrid')
sns.lmplot('total_deaths_per_million','human_development_index',data=vac, palette='coolwarm',size=6,aspect=1,fit_reg=False)
#Appears to be the same at lower HDI. Branches out to higher deaths at higher HDIs. 
```

Age

```python
sns.set_style('whitegrid')
sns.lmplot('total_deaths_per_million','aged_65_older',data=vac, palette='coolwarm',size=6,aspect=1,fit_reg=False)
#Concentration is at low values. Seems to increase linearly weekly at higher values. 
```

Smokers to Deaths

```python
smokerdata = vac.dropna(subset=['female_smokers'])
smokerdata = smokerdata.dropna(subset=['male_smokers'])

tsmokers = []

for index, row in smokerdata.iterrows(): 
  ts = row["female_smokers"] + row["male_smokers"]
  tsmokers.append(ts)

smokerdata['total_smokers'] = tsmokers
```

```python
sns.set_style('whitegrid')
sns.lmplot('total_deaths_per_million','total_smokers',data=smokerdata, palette='coolwarm',size=6,aspect=1,fit_reg=False)
#Concentrations seem to stay at 0 with many smokers, but higher data increases the death concentrations. 
```
