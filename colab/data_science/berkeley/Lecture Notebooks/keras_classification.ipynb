```python
url = 'https://media.githubusercontent.com/media/PacktPublishing/Hands-On-Gradient-Boosting-with-XGBoost-and-Scikit-learn/master/Chapter02/heart_disease.csv'

import pandas as pd

df = pd.read_csv(url)
df.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler

X = df.iloc[:, :-1]
y = df.iloc[:, -1]

X_train, X_test, y_train, y_test = train_test_split(X, y)

scaler = MinMaxScaler()

X_train = scaler.fit_transform(X_train)
X_test = scaler.fit_transform(X_test)
```

```python
from keras.models import Sequential
from keras.layers import Dense, Dropout
from keras.callbacks import EarlyStopping

num_cols = X_train.shape[1]

#init model
model = Sequential()

model.add(Dense(8, input_shape=(num_cols,),activation='relu'))


model.add(Dense(8, activation='relu'))

model.add(Dense(8, activation='relu'))

model.add(Dense(1, activation='sigmoid'))

print(model.summary())

model.compile(optimizer='adam', loss='binary_crossentropy', metrics='accuracy')

early_stopping = EarlyStopping(patience=15)

model.fit(X_train, y_train, epochs=1000, batch_size=20, validation_split=0.1,
          callbacks=[early_stopping])

model.evaluate(X_test, y_test)
```
