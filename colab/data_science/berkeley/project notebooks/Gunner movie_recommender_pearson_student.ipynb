# Problem 1: Separate Year and Title

```python
# 1a. Change the movie title to only a title without a year
movies = []
years = []
for index, row in df_movies.iterrows(): 
  movieyr = row['title'].split('(')
  movie = movieyr[0]
  yr = movieyr[1].replace(")","")
  movies.append(movie)
  years.append(yr)

df_movies['movie'] = movies
df_movies['year'] = years

df_movies.head()
# 1b. Create a new column with the year
```

# Merge DataFrames

Combine your DataFrames into one

```python
df = pd.merge(df_movies, df_ratings)
df.head()
```

# Add Count Column

The purpose of the count column is to filter your DataFrame so that highly correlated movies that your recommender displays are not necessarily from 1 user only who also gave a rare movie a 4 star rating. 

```python
df['count'] = df.groupby(["movie_id", "title"])["user_id"].transform("count")
df.head()
```

# Filter df

To get movies that are more likely correlated on account of more data, we can filter out movies with the least number of ratings. The filter is up to you.

```python
df = df[df['count']>=40]
```

# Pivot DataFrame

The pivoted DataFrame is the one you will use to make recommenders here and with most machine learning models.

```python
df_pivot = pd.pivot_table(df,values='rating',columns='title',index='user_id')
df_pivot.head()
```

```python
df_pivot.shape
```

# Pearson Correlation

Getting the Pearson Correlelation coefficient is a simple as using the method .corrwith on a column. The result is the correlation of all the other columns. These can be ordered top to bottom using the sort_values method 

```python
movie = df_pivot['Dune (1984)']
movie_correlations = df_pivot.corrwith(movie)
movie_correlations.sort_values(ascending=False)[:15]
```
