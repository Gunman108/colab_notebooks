```python
lables = []
for index, row in titanic.iterrows(): 
    if (row["Sex"]=='male'):
      lables.append(1)
    else:
      lables.append(0)
 
    

print(lables)

mdied = 0
fdied = 0
msurv = 0
fsurv = 0

for index, row in titanic.iterrows(): 
    if (row["Survived"]==0):
      if (row["Sex"]=='male'):
        mdied +=1
      else:
        fdied +=1
    else:
      if (row["Sex"]=='male'):
        msurv +=1
      else:
        fsurv +=1
 
    

print(lables)

print(mdied)
print(fdied)
print(msurv)
print(fsurv)
```

```python
titanic['lables'] = lables
```
