```python
import numpy as np

board =  np.array([['.','1','2','3','4'],['1','.','.','.','.'],['2','.','.','.','.'],['3','.','.','.','.'],['4','.','.','.','.']])

def check(board):
  end = False
  count = np.count_nonzero(board == '.')
  if count >= 16:
    end = True
  return end


class Pawn():
  def __init__(self, x,y):
    self.xpos = y
    self.ypos = x
    self.typ = 'p'
    board[self.xpos,self.ypos] = '♙'
  def update(self, p1, p2):
    board[self.xpos,self.ypos] = '♙'
    board[p1,p2] = '.'

    

class Rook():
  def __init__(self, x,y):
    self.xpos = y
    self.ypos = x
    self.typ = 'r'
    board[self.xpos,self.ypos] = '♖'
  def update(self, p1, p2):
    board[self.xpos,self.ypos] = '♖'
    board[p1,p2] = '.'
    

class Bishop():
  def __init__(self, x,y):
    self.xpos = y
    self.ypos = x
    self.typ = 'b'
    board[self.xpos,self.ypos] = '♗'
  def update(self, p1, p2):
    board[self.xpos,self.ypos] = '♗'
    board[p1,p2] = '.'


class Knight():
  def __init__(self, x,y):
    self.xpos = y
    self.ypos = x
    self.typ = 'k'
    board[self.xpos,self.ypos] = '♘'
  def update(self, p1, p2):
    board[self.xpos,self.ypos] = '♘'
    board[p1,p2] = '.'



class Queen():
  def __init__(self, x,y):
    self.xpos = y
    self.ypos = x
    self.typ = 'q'
    board[self.xpos,self.ypos] = '♕'
  def update(self, p1, p2):
    board[self.xpos,self.ypos] = '♕'
    board[p1,p2] = '.'

class King():
  def __init__(self, x,y):
    self.xpos = y
    self.ypos = x
    self.typ = 'K'
    board[self.xpos,self.ypos] = '♔'
  def update(self, p1, p2):
    board[self.xpos,self.ypos] = '♔'
    board[p1,p2] = '.'

pieces = []
game = True

p1 = Pawn(3,2)
k1 = Knight(4,1)
q1 = Rook(1,1)
pieces.append(p1)
pieces.append(k1)
pieces.append(q1)

while game == True:
  #inputs and rendering
  print(board)
  p1, p2 = input().split()
  p1 = int(p1)
  p2 = int(p2)
  m1,m2 = input().split()
  m2 = int(m2)
  m1 = int(m1)
  valid = False
  present = False
  loss = True

  #piece detection

  for piece in pieces:
    if piece.xpos == m1 and piece.ypos == m2:
          present = True

  for piece in pieces:
    if piece.xpos == p1 and piece.ypos == p2:

      #piece movement rules
      if piece.typ == 'p':
        if m1 == piece.xpos - 1:
          if m2 == piece.ypos - 1 or m2 == piece.ypos + 1:
            valid = True
        

      if piece.typ == 'r':
        if m1 == piece.xpos or m2 == piece.ypos:
          valid = True

        

      if piece.typ == 'k':
        if m1 == piece.xpos + 2 or m1 == piece.xpos -2:
          if m2 == piece.ypos - 1 or m2 == piece.ypos + 1:
            valid = True
        if m2 == piece.ypos + 2 or m2 == piece.ypos -2:
          if m2 == piece.xpos - 1 or m2 == piece.xpos + 1:
            valid = True


      if piece.typ == 'b':
        if abs(m1 - p1) == abs(m2 - p2):
          valid = True
        
      
      if piece.typ == 'q':
        if abs(m1 - p1) == abs(m2 - p2):
          valid = True
        if m1 == piece.xpos or m2 == piece.ypos:
          valid = True
      
      if piece.typ == 'K':
        if abs(m1 - p1) == 1 or abs(m2-p2) == 1:
          valid = True

      if m1 == 0 or m2 == 0:
        valid = False

      #Code to check for loss
      for piece in pieces:
        if piece.tpy == 'p':
          if board[piece.xpos-1,piece.ypos-1] != '.' or board[piece.xpos-1,piece.ypos+1] != '.':
            loss = False
        if piece.tpy == 'r':
          if board[piece.xpos,:].any != '.' or board[:,piece.ypos].any != '.':
            loss = False
        if piece.tpy == 'b':
          for i in range(4):
            if board[piece.xpos+-i,piece.ypos+-i] != '.':
              loss = False
        if piece.typ == 'k':
          if board[piece.xpos +- 2,board.ypos +- 1] != '.':
            loss = False
          if board[piece.xpos +- 1, board.ypos +- 2] != '.':
            loss = False

      
        if piece.tpy == 'q':
          for i in range(4):
            if board[piece.xpos+-i,piece.ypos+-i] != '.':
                loss = False
          if board[piece.xpos,:].any != '.' or board[:,piece.ypos].any != '.':
            loss = False

        if piece.tpy == 'K':
          if board[piece.xpos+-1,piece.ypos+-1] != '.' or board[piece.xpos,piece.ypos+-1] != '.' or board[piece.xpos+-1,piece.ypos] != '.':
            loss = False

      #piece movement / updating
      if valid == True and present == True:
        print("good move")
        piece.xpos = m1
        piece.ypos = m2
        piece.update(p1,p2)
        if check(board) == True:
          print(board)
          print("WIN")
          game = False
          
      else:
        print("INVALID MOVE")

      if loss == True:
        print(board)
        game = False
        print("LOOSE")



```
