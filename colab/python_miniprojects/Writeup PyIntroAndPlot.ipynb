Imports Libraries for Data Visualization and File Uploads

```python
import pandas as pd
#import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

import io
from google.colab import files
uploaded = files.upload()
```

Creates dataset reg based off of the file that we read

```python
reg = pd.read_csv(io.BytesIO(uploaded['regress.csv']))
```

Head function prints our first five in our dataset

```python
reg.head()
```

Info function determines the type of variables, amount of nulls, memory usage, and a few other useful info for data preprocessing

```python
reg.info()
```

Creates a histogram for each column in reg that tells us at what datapoints our data is most dense

```python
reg.hist()
```

Describe gives us useful things like count, mean, min, and percentages

```python
reg.describe()
```

Plots both a histogram and scatterplot next to each other to compare density of our data on the histogram with actual datapoints on the scatter

```python
sns.pairplot(reg)
```

Uses Linear Regression on a scatterplot to determine the best fit linear relationship in our data. 

```python
sns.lmplot(x='SAT',y='GPA',data=reg)
```

Creates a more simplified graph showing the curve of the density in our data. (Resembles the Bell Curve, the most common density curve)

```python
sns.distplot(reg['SAT'])
```
