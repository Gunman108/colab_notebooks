```python
import requests

list = []

class Character(object):
    #Init function does the same thing as a constructor in java
    def __init__(self, name, height, mass, hair_color, skin_color,eye_color,birth_year,gender):
      self.name = name
      self.height = height
      self.mass = mass
      self.hair_color = hair_color
      self.skin_color = skin_color
      self.eye_color = eye_color
      self.birth_year = birth_year
      self.gender = gender

    def prnt(self): # print function - NEEDS TO TAKE IN SELF
      print()
      print("Name:",self.name) 
      print("Height:",self.height,"cm")
      print("Mass:",self.mass,"kg")
      print("Hair Color:",self.hair_color)
      print("Skin Color:",self.skin_color)
      print("Eye Color:",self.eye_color)
      print("Birth Year:",self.birth_year)
      print("Gender:",self.gender)

for i in range(10):
  url = f'https://swapi.dev/api/people/{i+1}/' #Adding variables to strings has slightly different syntax in python. 
  
  json_data = requests.get(url).json() #Makes a request to the url and outputs a dictionary



#Input statements
  n = json_data['name'] #json_data is output as a dictionary. We can access it through square brackets and a label
  h = json_data['height']
  m = json_data['mass']
  hc = json_data['hair_color']
  s = json_data['skin_color']
  e = json_data['eye_color']
  b = json_data['birth_year']
  g = json_data['gender']


  
  list.append(Character(n,h,m,hc,s,e,b,g)) #Creates and adds new object to a list of objects

for obj in list: #This loop iterates through all objects in lists and calls the print function for each
  obj.prnt()
```
