Lists and tuples

```python
mylist = ["a","b","c","d"]
mylist2 = ["f","g","h","i"]

mylist.append("e")
mylist.insert(0, "alphabet:")
mylist.extend(mylist2)

mylist3 = [i for i in mylist if "a" in i]

for i in mylist:
  print(i)
  mylist.remove(i)

print(mylist)
print(mylist2)
print(mylist3)
```
