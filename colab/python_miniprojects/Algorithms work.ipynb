Doors Problem

```python
def toggle(door):
  return not door
doors = []
for i in range(100):
  doors.append(False)
```

```python
for i in range(101):
  for x in range(i, 101, i):
    if doors[x] % (i+1) == 0:
      doors[x] = not doors[x]
print(doors)
```

FizzBuzz

```python
for i in range(1, 100):
  if i % 3 == 0 and i % 5 == 0:
    print('fizzbuzz')
  elif i % 3 == 0:
    print('fizz')
  elif i % 5 == 0:
    print('buzz')
  else:
    print(i)
```

```python
for i in range(1, 100): #interested to see which approach is quicker
  fb = ''
  if i % 3 == 0:
    fb += 'fizz'
  if i % 5 == 0:
    fb += 'buzz'
  if i % 3 != 0 and i % 5 != 0:
    fb += str(i)
  print(fb)
```
