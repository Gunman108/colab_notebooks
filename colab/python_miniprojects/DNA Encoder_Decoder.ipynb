```python
dna_strand = "TACTCCTTT"
rna_strand = ""

#TTT TAT

#AGG AAA

#AUA AUU

def transcribe(input_dna_strand):
  result = ""
  for character in input_dna_strand:
    if character is 'A':
      result = result + 'U'
    elif character is 'C':
      result = result + 'G'
    elif character is 'T':
      result = result + 'A'
    elif character is 'G':
      result = result + 'C'
  return result

rna_strand = transcribe(dna_strand)
print(rna_strand)
```
