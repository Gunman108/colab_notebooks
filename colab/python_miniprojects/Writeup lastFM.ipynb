Takes in API and Params. Then prints the data recieved

```
import requests

url = 'http://ws.audioscrobbler.com/2.0/' #The URL of our API

params = { #These are the perameters we take in
	'method': 'artist.gettoptracks', #Only takes in top tracks
	'artist': 'beyonce', #Spesifies which artist
	'api_key': 'b496f5382c08d1be6cd606f9639e522e', #We need an api key because our api is private. This will act as our "password"
	'format': 'json' #Spesifies which format we want (in this case is json)
}

req = requests.get(url, params=params)
res = req.json()

print(res) #Prints the dictionary
```
