Imports the api and Prints and sorts different types of Tracks by length

---



```
import requests

url = 'https://itunes.apple.com/search' #Takes in the url of the api

data = { #Here it takes in the name of the artist and the amount of things you want to pull
	'term': 'beyonce',
	'limit': '15'
}

req = requests.get(url, params=data) #creates variable req with perameters of the url and data(term and limit)

res = req.json() #creates dictionary res equal to the json request

print(res) #Prints the dictionary

print("----- List of Tracks: -----")

tracks = [r.get('trackName') for r in res.get('results')] # List comprehension

for r in res.get("results"): #creates a list with all tracknames in dictionary
	tracks.append(r.get("trackName"))


print(tracks) #prints the list

print("\n----- Short Tracks: -----") 
short_tracks = [r.get('trackName') for r in res.get('results') if r.get('trackTimeMillis') < 250000] # List comprehension with a condition (here it is the length of the song)
print('\n'.join(short_tracks)) # Join list with new line for better view 

print("\n----- Songs by Length: -----")
length_tracks = [(r.get('trackName'), r.get('trackTimeMillis')) for r in res.get('results')]
sorted_tracks = sorted(length_tracks, key=lambda k: k[1]) # Sort by 2nd item in the tuple (length) This list is sorted by ascending order by legnth
print('\n'.join([t[0] + ' - ' + str(t[1]) for t in sorted_tracks])) #prints the length and name of the track


```

Takes the API again and seaches for Queen songs, then prints them in order of release date

```
import requests

url = 'https://itunes.apple.com/search'
params = { #takes in queen as search term (if no amount of terms is given it will take in all)
	'term': 'Queen'
}

req = requests.get(url, params=params) #see beyonce code
res = req.json()

for i in res.get('results'):
  print(i.get('trackName') + " " + str(i.get('releaseDate'))) #Prints all track names with release dates (prints in order of release by default)

```
